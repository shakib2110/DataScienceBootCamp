```python
import math

str1 = input().split(" ")
a, b, c = str1

a = float(a)
b = float(b)
c = float(c)

d =((b*b)-(4*a*c))

if ((d>=0) and a!=0):
    r1 = ((-b) + math.sqrt(d))/ (2*a)
    r2 = ((-b) - math.sqrt(d))/ (2*a)
    print("R1 = {:.5f}\nR2 = {:.5f}" .format(r1,r2))
    
else:
    print("Impossivel calcular")
```

```python
str1 = input().split(" ")

a, b, c = str1

a = float(a)
b = float(b)
c = float(c)

if (a<b):
    a, b = b, a

if (b<c):
    b, c = c, b

if (a<b):
    a, b = b, a
    
if (a >= (b + c)):
    print("NAO FORMA TRIANGULO")

else:
    if ((a*a) == ((b*b) + (c*c))):
        print("TRIANGULO RETANGULO")

    if ( (a*a) > ((b*b) + (c*c))):
        print("TRIANGULO OBTUSANGULO")

    if ( (a*a) < ((b*b) + (c*c))):
        print("TRIANGULO ACUTANGULO")
        
    if  ((a==b and a!=c) or (b==c and b!=a) or (c==a and c!=b)):
        print("TRIANGULO ISOSCELES")
        
    if (a==b and b==c and c==a):
        print("TRIANGULO EQUILATERO")
```

Game Time with Minutes

```python
str1 = input().split(" ")

h1, m1, h2, m2= str1

h1 = int(h1)
m1 = int(m1)
h2 = int(h2)
m2 = int(m2)

tem1 = (h1*60) + m1
tem2 = (h2*60) + m2

if (tem1==tem2):
    x = 24
    print("O JOGO DUROU {} HORA(S) E 0 MINUTO(S)" .format(x))
    
elif (tem1>tem2):
    x = (24*60) - (tem1 - tem2)
    y = x/60
    z = x%60
    print("O JOGO DUROU {} HORA(S) E {} MINUTO(S)" .format(int(y), int(z)))
    
elif (tem1 < tem2):
    x = tem2 - tem1
    y = x/60
    z = x%60
    print("O JOGO DUROU {} HORA(S) E {} MINUTO(S)" .format(int(y), int(z)))
```
